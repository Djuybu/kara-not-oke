```python
import os
import glob
import re
```

```python
INPUT_DIR = '../media_files/separated_audio/train'

VOCAL_DIR = INPUT_DIR + '/vocals'
INSTRUMENTAL_DIR = INPUT_DIR + '/instrumentals'
```

```python
for entry in glob.glob(os.path.join(INPUT_DIR, '*.wav')):
    filename = os.path.basename(entry)
    
    # 1. Tách phần tên và phần mở rộng
    name_part, extension = os.path.splitext(filename)
    
    # 2. Xóa phần từ '.stem' cho đến hết tên (trước phần mở rộng)
    clean_name = name_part.split('.stem')[0]
    
    # 3. Xóa cụm (Vocals) và (Instrumental)
    clean_name = clean_name.replace('(Vocals)', '').replace('(Instrumental)', '')
    
    # 4. Thay thế khoảng trắng bằng dấu gạch ngang
    # \s+ đại diện cho 1 hoặc nhiều khoảng trắng liên tiếp
    clean_name = re.sub(r'\s+', '-', clean_name).strip('-')
    
    # Tên file mới hoàn chỉnh
    new_filename = f"{clean_name}{extension}"
    
    # Phân loại và di chuyển file
    if '(Vocals)' in filename:
        dest_path = os.path.join(VOCAL_DIR, new_filename)
        os.rename(entry, dest_path)
        print(f"Moved Vocal: {new_filename}")
    elif '(Instrumental)' in filename:
        dest_path = os.path.join(INSTRUMENTAL_DIR, new_filename)
        os.rename(entry, dest_path)
        print(f"Moved Instrumental: {new_filename}")
```
